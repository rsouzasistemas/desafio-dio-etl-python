# **Extract**

```python
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
from matplotlib import rcParams

# Extração
electricVehiclePopulation = pd.read_csv('rows.csv', encoding='ISO-8859-1')
```

# **Transform**

```python
#Transformação
dataInfo = electricVehiclePopulation.shape
dataInfo = f"O DataFrame possui {dataInfo[0]} linhas e {dataInfo[1]} colunas"

makers = electricVehiclePopulation['Make'].unique()
years = electricVehiclePopulation['Model Year'].unique()
countMakers = len(electricVehiclePopulation['Make'].unique())
firstYear = electricVehiclePopulation['Model Year'].min()
lastYear = electricVehiclePopulation['Model Year'].max()

groupByYear = electricVehiclePopulation.groupby(electricVehiclePopulation['Model Year'])['Make'].count()
```

# **Load**

```python
#Exibição
%matplotlib inline
rcParams['figure.figsize'] = 15, 4

electricVehiclePopulation.groupby(electricVehiclePopulation['Make'])['Model Year'].count().plot.bar()

plt.title("Fabricados por marca")
plt.xlabel("Marca")
plt.ylabel("Quantidade")
plt.show()

electricVehiclePopulation.groupby(electricVehiclePopulation['Model Year'])['Make'].count().plot()

plt.title("Fabricados por ano")
plt.xlabel("Ano")
plt.ylabel("Quantidade")
plt.show()
```
